```python
import struct
import numpy as np

# --- Method 1: using struct ---
def bytes_to_vector_struct(b: bytes):
    n = len(b) // 4  # float32 = 4 bytes
    return list(struct.unpack(f"{n}f", b))


# --- Method 2: using numpy (recommended) ---
def bytes_to_vector_numpy(b: bytes):
    return np.frombuffer(b, dtype=np.float32)
```
